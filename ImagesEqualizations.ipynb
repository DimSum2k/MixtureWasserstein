# Grayscale Histogram Equalization

```python
# Import a first image
from scipy import misc
import numpy as np
f = misc.face()

import matplotlib.pyplot as plt
plt.imshow(f)
plt.show()
```

```python
# convert to rgb
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
f2 = rgb2gray(f)

n, m  = 768, 1024
f2.shape
```

```python
plt.imshow(f2)
plt.show()
```

```python
# plot histogram with 255 bins
b, bins, patches = plt.hist(f2.flatten(), 255)
plt.xlim([0,255])
plt.show()
```

```python
# Apply transformation
f2_flat = f2.flatten()
f2_flat[np.argsort(f2_flat)] =  np.linspace(0,1,n*m)
#f3 = (255 * (1.0 - f2_flat)).reshape(n,m)
f3 = (255 * f2_flat).reshape(n,m)
```

```python
plt.imshow(f3)
plt.show()
```

```python
# plot histogram with 255 bins
b, bins, patches = plt.hist(f3.flatten(), 255)
plt.xlim([0,255])
plt.show()
```

```python
# Let's try to interpolate
for alpha in np.linspace(0,1,10):
    f_alpha = f2*(1-alpha) + f3*alpha
    b, bins, patches = plt.hist(f_alpha.flatten(), 255)
    plt.xlim([0,255])
    plt.show()
    plt.imshow(f_alpha)
    plt.show()
```
